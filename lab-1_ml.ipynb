Напишите функцию linan(), которая будет решать систему линейных уравнений вида:

\begin{matrix} a_{11}x+a_{12}y=b_{1}&(1) \\ a_{21}x+a_{22}y=b_{2} &(2) \end{matrix}
​
На вход программа должна принимать 2 строки.

Каждая строка содержит 3 действительных числа через пробел: коэффициенты a и b.

На выходе программа должна выдавать значения x и y в одну строку через пробел.


Примечание 1. Для разделения строки на отдельные элементы используйте функцию split()

Примечание 2. В качестве разделителя между целой и дробной частью используйте точку, например 0.2

````python
# enter you code

def linan(s1 : str, s2 : str):
    arr2d = np.array([s1.split(), s2.split()], dtype=float)
    A, b = np.split(arr2d, [2], axis=1)
    x, y = linalg.solve(A, b).ravel()
    return f'{x}, {y}'
````

````python
s1 = '1 -1 7'
s2 = '3 2 16'
linan(s1, s2)
````

Нужно сделать без использования  циклов.

```
# ожидаемый результат для аргумента 5

```



Обратите внимание, что ожидается массив numpy

````python
def decorate_matrix(n : int):
    res = np.zeros((n - 2, n - 2))
    res = np.pad(res, 1, constant_values=1)
    return res
````

# Задание 3

1) реализуйте 2 функции:

$f(x) = (x+a)^2 - b$

$g(x) = |f(x)|$

2) Найдите аргументы, при которых достигается минимум функций $f(x)$ и $g(x)$

Пример.
Входные значение: $a=3$ и $b=4$

Выходные значения:

`-3.0 -1.0`

````python
def function_1(a : float, b : float):
    return lambda x: (x + a)**2 - b

def function_2(f1 : callable):
    return lambda x: abs(f1(x))
````

Производная первой функции
\begin{equation}
    f'_1(x) = \frac{d}{dx}\left((x + a)^2 - b\right) = 2(x + a)
\end{equation}
Производная равна $0$ в точке $x = -a$, поэтому данная точка является критической. Слева от нее производная отрицательна, справа полложительна, поэтому данная точка $x = -a$ является точкой минимума


Производная второй функции (используя $\frac{d}{dx}|x| = \frac{d}{dx}\sqrt {x^2} = \frac{x}{|x|}$)
\begin{equation}
    f'_2(x) = \frac{d}{dx}\left(\left|(x + a)^2 - b\right|\right) = \frac{(x + a)^2 - b}{|(x + a)^2 - b|} \cdot \frac{d}{dx}\left((x + a)^2 - b\right) = \frac{(x + a)^2 - b}{|(x + a)^2 - b|} \cdot 2(x + a)
\end{equation}

\begin{equation}
    f'_2(x) = 
    \begin{cases}
        2(x + a) & \quad (x + a)^2 > b \\
        -2(x + a) & \quad (x + a)^2 < b
    \end{cases}
\end{equation}

Производная равна $0$ в одной точке $x = -a$, эта точка критическая

Если $b > 0$, производная терпит разрыв в критических точках
\begin{gather}
    (x + a)^2 - b = 0 \\
    x^2 + 2ax + (a^2 - b) = 0 \\
    x = \frac{-2a \pm \sqrt{((2a)^2 - 4(a^2 - b))}}{2} = -a \pm \sqrt b
\end{gather}
Слева от $x = -a - \sqrt b$ производная отрицательна, справа положительна, данная точка является точкой минимума. Аналогично для точки $x = -a + \sqrt b$. Точка $x = -a$ не является минимумом, так как слева от нее производная положительна, справа отрицательна

Если $b \le 0$ разрывов нет, единственной критической точкой является $x = -a$. Слева от нее производная отрицательная, справа - положительная, значит эта точка является точкой минимума

Итого, если $b > 0$ имеются два минимума $x = -a - \sqrt b$ и $x = -a + \sqrt b$, если $b \le 0$ минимум в $x = -a$

````python
def minimize(a, b):
    if (b > 0):
        return (-a, ), (-a - np.sqrt(b), -a + np.sqrt(b))
    else:
        return (-a, ), (-a, )
````

````python
a = 5
b = 12

f1 = function_1(a, b)
f2 = function_2(f1)

min_f1, min_f2 = minimize(a, b)
print(min_f1, min_f2)
````

# Задание 5

1000 человек бросают монетку. Каждый бросил 10 раз подряд.

 1. Какова вероятность того, что хотя бы у одного человека монетка выпала 10 раз подряд той же самой стороной (орлом или решкой)?
 2. Какова вероятность того, что хотя бы у одного человека монетка выпала 8 раз подряд той же самой стороной (орлом или решкой)?
 3. Какова вероятность того, что ровно у троих человек из 1000 монетка выпала 10 раз подряд той же самой стороной (орлом или решкой)?

Использую биномиальное распределение, где true - падение решки, false - падение орла, вероятность успеха либо провала $p=0.5$, количество бросков $n=10$, количество экспериментов по броскам $size=1000$

````python
p = 0.5
n = 10
size = 1000

sim_size = 200
````

````python
from scipy.stats import binom

cnt = 0
for i in range(sim_size):
    if np.isin(10, binom.rvs(n, p, size=size)) or np.isin(0, binom.rvs(n, p, size=size)):
        cnt += 1
print(cnt / sim_size)
````

````python
def is_slice(l, s):
    n = len(l)
    m = len(s)
 
    for i in range(n - m + 1):
        for j in range(m):
            if l[i + j] != s[j]:
                break
        else: 
            return True
    return False
````

````python
cnt = 0
ones = np.ones(10)
zeros = np.zeros(10)
for i in range(50):
    for j in range(size):
        data = binom.rvs(1, p, size=n)
        if is_slice(data, ones) or is_slice(data, zeros):
            cnt += 1
            break
print(cnt / 50)
````

````python
cnt = 0
ones = np.ones(10)
zeros = np.zeros(10)
for i in range(sim_size):
    inner_cnt = 0
    for j in range(size):
        data = binom.rvs(1, p, size=n)
        if is_slice(data, ones) or is_slice(data, zeros):
            inner_cnt += 1
    if inner_cnt == 3:
        cnt += 1
print(cnt / sim_size)
````
